# **Exploratory Data Analysis :**

**STEPS OF EDA :**
1. **Understanding the Dataset**
- Data types (numerical, categorical, text, datetime, etc.)
- Shape of the dataset (rows & columns)
- Summary statistics (mean, median, mode, standard deviation)
- Checking for missing values
2. **Data Cleaning & Preprocessing**
- Handling missing values (imputation, deletion)
- Dealing with duplicates
Outlier detection and treatment
- Encoding categorical variables (One-Hot Encoding, Label Encoding)
- Normalization & Standardization
3. **Univariate Analysis (Single Variable**)
- Histograms & KDE plots
- Box plots & Violin plots
- Count plots for categorical variables
- Descriptive statistics (mean, median, variance, skewness, kurtosis)
4. **Bivariate & Multivariate Analysis**
- Scatter plots (for numerical variables)
- Pair plots & Correlation heatmaps
- Cross-tabulation & Chi-square test (for categorical variables)
- Boxplots & Violin plots (comparing numerical vs categorical data)
- Pivot tables & Groupby operations
5. **Feature Engineering & Transformation**
- Creating new features from existing data
- Log transformation & Power transformation (handling skewed data)
- Binning numerical values into categories
- Feature selection techniques
6. **Correlation & Multicollinearity**
- Pearson & Spearman correlation
- Heatmaps for correlation visualization
- Variance Inflation Factor (VIF) to detect multicollinearity
7. **Dimensionality Reduction**
- PCA (Principal Component Analysis)
- t-SNE & UMAP for visualization
8. **Data Visualization**
- Bar charts, Pie charts, Line plots
- Heatmaps & Pair plots
- Treemaps & Sunburst charts (for hierarchical data)

**Steps :**
1. Load the Data
2. Understand the Data
3. Univariate Analysis
4. Bivariate & Multivariate Analysis
5. Handling Missing Data
6. Outlier Detection
7. Feature Engineering
